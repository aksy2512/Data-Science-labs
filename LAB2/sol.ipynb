```python
of = pd.read_csv('landslide_data3_original.csv')
of
```

```python
new_df=df.copy()
for i in range(len(col)):
    if(i>0):
     new_df[col[i]].fillna(np.mean(df[col[i]]),inplace=True)
new_df
```

```python
new_df.describe()
```

```python
for i in range(len(col)):
    if(i>0):
     print(f'The mean for {col[i]} in original file is {np.mean(of[col[i]])} and for filled file is {np.mean(new_df[col[i]])}')
     print(f'The median for {col[i]} in original file is {np.median(of[col[i]])} and for filled file is {np.median(new_df[col[i]])}')
     print(f'The mode for {col[i]} in original file is {stats.mode(of[col[i]])[0]} and for filled file is {stats.mode(new_df[col[i]])[0]}')
     print(f'The stddev for {col[i]} in original file is {np.std(of[col[i]])} and for filled file is {np.std(new_df[col[i]])}')
     print("---------------------------------------------------------------------------------------------------------------------------------------------")

```

```python

for i in range(len(col)):
    sumq=0
    c=0
    if(i>0):
     for index,row in df.iterrows():
       if(pd.isnull(df[col[i]].loc[index])):
            sumq = sumq + (of[col[i]].loc[index]-new_df[col[i]].loc[index])**2
            c=c+1   
     rsme = (sumq/c)**0.5
     plt.bar(col[i],rsme)
     plt.xticks(col,rotation='vertical')

     
     print(f'The root mean square error (RMSE) between the original and replaced values for {col[i]} is {rsme}') 
plt.show()    
     
            


```
